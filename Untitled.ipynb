```python
import cv2 as cv
import mediapipe as mp
import numpy as np 
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
#Video Feed
cap = cv.VideoCapture(0)
#setup Mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret,frame = cap.read()   #ret : return value , frame :give image from webcam
        
        #Recolor image to RGB
        image = cv.cvtColor(frame,cv.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        #Make Detection ,store our detection into results
        results = pose.process(image)
        
        
        #Recolor image to BGR
        image.flags.writeable = True
        image = cv.cvtColor(image, cv.COLOR_RGB2BGR)
        
        #Render Detection 
        mp_drawing.draw_landmarks(image,results.pose_landmarks,mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245,117,66),thickness=2,circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245,66,230),thickness=2,circle_radius=2),
                                  )
        
        print(results)
        
        
        
        cv.imshow("Mediapipe Feed", image)   #pop up for showing image 
        if cv.waitKey(10)& 0xFF == ord('q'):
            break

    cap.release()   
    cv.destroyAllWindows()
```

```python
def calculate_angle(a,b,c):
    a = np.array(a) #first
    b = np.array(b) #mid
    c = np.array(c) #end
    
    radians = np.arctan2(c[1]-b[1],c[0]-b[0])-np.arctan2(a[1]-b[1],a[0]-b[0])
    angle = np.abs(radians *180.0/np.pi)
    
    if angle > 180.0:
        angle = 360-angle
        
    return angle 

    
```

```python
#Video Feed
cap = cv.VideoCapture(0)

#Curl counter variables
countr = 0
stage = None


#setup Mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret,frame = cap.read()   #ret : return value , frame :give image from webcam
        
        #Recolor image to RGB
        image = cv.cvtColor(frame,cv.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        #Make Detection ,store our detection into results
        results = pose.process(image)
        
        
        #Recolor image to BGR
        image.flags.writeable = True
        image = cv.cvtColor(image, cv.COLOR_RGB2BGR)
        
        
        #Extract Landmarks
        try:
            landmarks = results.pose_landmarks.landmark
            
            #Get coordinates
            shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
            
            #Calculate Angle
            angle = calculate_angle(shoulder,elbow,wrist)
            
            
            #Visualize angle
            cv.putText(image,str(angle),
                       tuple(np.multiply(elbow,[640,480]).astype(int)),
                             cv.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv.LINE_AA
                       
                       )
            
            #Curl Counter Logic
            if angle>160:
                stage = 'down'
            if angle<30 and stage == 'down':
                stage = 'up'
                countr += 1



        except:
            pass
        
        #Render curl counter
        #setup status box
        cv.rectangle(image,(0,0),(225,73),(245,117,16),-1)
        
        #Rep data
        cv.putText(image,'Reps',(15,12),
                   cv.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),1,cv.LINE_AA)
        cv.putText(image,str(countr),(10,60),
                   cv.FONT_HERSHEY_SIMPLEX,2,(255,255,255),2,cv.LINE_AA)
        #Stage data
        cv.putText(image,'Stage',(65,12),
                   cv.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),1,cv.LINE_AA)
        cv.putText(image,stage,(70,60),
                   cv.FONT_HERSHEY_SIMPLEX,2,(255,255,255),2,cv.LINE_AA)
        
        
        #Render Detection 
        mp_drawing.draw_landmarks(image,results.pose_landmarks,mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245,117,66),thickness=2,circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245,66,230),thickness=2,circle_radius=2),
                                  )
        
        
        
        
        cv.imshow("Mediapipe Feed", image)   #pop up for showing image 
        if cv.waitKey(10)& 0xFF == ord('q'):
            break

    cap.release()   
    cv.destroyAllWindows()
```
